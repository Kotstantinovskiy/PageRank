```python
urls = {}
urls_file = open("urls_idx.txt", "r")
for line in urls_file:
    if "lenta.ru" in line.split("\t")[1][:-1]:
        urls[int(line.split("\t")[0])] = line.split("\t")[1][:-1]
urls_file.close()
```

```python
page_rank_file = open("pageRankOut.txt", "r")
indx_pagerank = {}
for line in page_rank_file:
    indx_pagerank[int(line.split("\t")[0])] = float(line.split("\t")[1])
page_rank_file.close()
```

```python
auth_file = open("authOut.txt", "r")
indx_auth = {}
for line in auth_file:
    indx_auth[int(line.split("\t")[0])] = float(line.split("\t")[2])
auth_file.close()
```

```python
hub_file = open("hubOut.txt", "r")
indx_hub = {}
for line in hub_file:
    indx_hub[int(line.split("\t")[0])] = float(line.split("\t")[2])
hub_file.close()
```
